```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Simulating M&A data
np.random.seed(0)  # For reproducibility

# Simulate data for 10 years
years = pd.date_range(start='2013-01-01', end='2022-12-31', freq='Y')

# Simulating the number of deals and deal values
data = {
    'Year': years.year,
    'Number_of_Deals': np.random.randint(5, 20, size=len(years)),
    'Total_Deal_Value': np.random.randint(1_000_000, 100_000_000, size=len(years))
}

ma_data = pd.DataFrame(data)
```

```python
# Calculate additional metrics
ma_data['Average_Deal_Size'] = ma_data['Total_Deal_Value'] / ma_data['Number_of_Deals']

# Summary statistics
summary_stats = ma_data.describe()
print(summary_stats)
```

```python
# Plotting the number of deals over time
plt.figure(figsize=(12, 6))
sns.barplot(x='Year', y='Number_of_Deals', data=ma_data, palette='viridis')
plt.title('Number of M&A Deals Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Deals')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

# Plotting total deal value over time
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Total_Deal_Value', data=ma_data, marker='o', color='orange')
plt.title('Total Deal Value Over Time')
plt.xlabel('Year')
plt.ylabel('Total Deal Value ($)')
plt.xticks(rotation=45)
plt.grid()
plt.show()

# Plotting average deal size over time
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Average_Deal_Size', data=ma_data, marker='o', color='blue')
plt.title('Average Deal Size Over Time')
plt.xlabel('Year')
plt.ylabel('Average Deal Size ($)')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
# Calculate the growth rate of deals and deal value
ma_data['Deal_Growth'] = ma_data['Number_of_Deals'].pct_change() * 100
ma_data['Value_Growth'] = ma_data['Total_Deal_Value'].pct_change() * 100

# Display the growth metrics
print(ma_data[['Year', 'Deal_Growth', 'Value_Growth']])
```
